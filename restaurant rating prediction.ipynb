```python

from sklearn.preprocessing import LabelEncoder
# List of categorical columns
categorical_columns = ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 
                       'Currency', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 
                       'Switch to order menu', 'Rating color', 'Rating text']

# Convert all categorical columns to string type
for column in categorical_columns:
    df[column] = df[column].astype(str)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to each categorical column
for column in categorical_columns:
    df[column] = label_encoder.fit_transform(df[column])

# Check the transformed data
print(df.head())


```

```python
from sklearn.preprocessing import LabelEncoder

# Define the dependent variable (y)
y = df['Aggregate rating']

# Define the independent variables (X)
X = df.drop(columns=['Aggregate rating'])

# List of categorical columns in X
categorical_columns = ['Restaurant Name', 'City', 'Address', 'Locality', 'Locality Verbose', 'Cuisines', 
                       'Currency', 'Has Table booking', 'Has Online delivery', 'Is delivering now', 
                       'Switch to order menu', 'Rating color', 'Rating text']

# Convert all categorical columns in X to string type
for column in categorical_columns:
    X[column] = X[column].astype(str)

# Initialize LabelEncoder
label_encoder = LabelEncoder()

# Apply LabelEncoder to each categorical column in X
for column in categorical_columns:
    X[column] = label_encoder.fit_transform(X[column])


```

```python
print(X.head())
```

```python
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

# linear regression

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression() 
reg.fit(X_train,y_train)
```

```python
y_pred = reg.predict(X_test)

from sklearn.metrics import r2_score    
r2 = r2_score(y_test,y_pred)
print(r2)
```

```python
plt.figure(figsize=(9, 5))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Aggregate Rating')
plt.ylabel('Predicted Aggregate Rating')
plt.title('Actual vs Predicted Aggregate Rating')
plt.show()
```
